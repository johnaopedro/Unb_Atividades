```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mediaVoo539k.csv')
df2 = pd.read_csv('mediaVoo.csv')
display(df.loc[:,['Airline','AverageDelay']], df2.loc[:,['Airline','AverageDelay']])
```

```python
fig, graficoMedia = plt.subplots(1,2, figsize=(18,4))

graficoMedia[0].plot(df['Airline'].tolist(), df['AverageDelay'].tolist(), '-bo')
graficoMedia[0].grid()
graficoMedia[0].set_title('Média de atrasos [%d voos]' %df['amountFlight'].sum())

graficoMedia[1].plot(df2['Airline'].tolist(), df2['AverageDelay'].tolist(), '-bo')
graficoMedia[1].grid()
graficoMedia[1].set_title(f'Média de atrasos [%d voos]' %df2['amountFlight'].sum())

for i in graficoMedia.flat:
    i.set(xlabel='Companhias Aéreas', ylabel='Média de atrasos')
plt.show()
```
